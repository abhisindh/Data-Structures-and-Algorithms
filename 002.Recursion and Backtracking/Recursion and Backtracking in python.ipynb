# Factorial function

### C code

```c
// calculates factorial of a positive integer

int Fact(int n){
    if (n==1){ // base case: fact of 0 or 1 is 1
        return 1;
    }else if (n==0){
        return 1;
    }else{ //recursive case: multiply n by (n-1) factorial
        return n*Fact(n-1);
    }
}
```

### Python code

````python
# calculates factorial of a positive integer
def Fact(n):
    if n==1 or n==0: # base case: fact of 0 or 1 is 1
        return 1
    else: #recursive case: multiply n by (n-1) factorial
        return n*Fact(n-1)
if __name__ == '__main__':
    n = int(input("Enter a number : "))
    print(f"factorial of {n} is {Fact(n)}")
````

# Print numbers 1 to n backward

### C code

```c
// Print numbers 1 to n backward
int Print(int n){
    if (n==0){ // base case
        return 0;
    }else{ // recursive case
        printf("%d \n",n);
        return Print(n-1);
    }
}
```

### Python code

````python
# Print numbers 1 to n backward

def Print(n):
    if n==0:
        return 0
    else:
        print(n)
        Print(n-1)
Print(5)
````
